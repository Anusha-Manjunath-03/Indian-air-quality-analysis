```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
df = pd.DataFrame(data)
```

```python
#Check missing values
df.head()
```

```python
#Check for duplicate values
df[df.duplicated()]
```

```python
#Check for correlation
data.corr(method = 'pearson')
```

```python
#Check for percentage of NaN values of the columns
df.isnull().sum()/len(df)*100
```

```python
#dropping the columns that are not used for the analysis and columns with more NaN values
converted_df = df.drop(['pm2_5','stn_code','sampling_date','agency','location_monitoring_station','pm2_5'],axis = 1)
```

```python
converted_df
```

```python
#dropping of rows that has few NaN values
converted_df = converted_df.dropna(axis = 0,subset = ['type'])
converted_df = converted_df.dropna(axis = 0,subset = ['so2'])
converted_df = converted_df.dropna(axis = 0,subset = ['location'])
```

```python
converted_df
```

```python
#replacing the NaN values with the mean of their respective columns
converted_df[['rspm','spm']] = converted_df[['rspm','spm']].fillna(value = converted_df.mean())
```

```python
converted_df.head()
```

```python
#preprocessed data to excel
converted_df.to_excel('preprocessed.xlsx')
```

```python
dataset=pd.read_excel("preprocessed.xlsx")
```

```python
dataset
```

```python
#Categorical variables to dummy variables
df_preprocessed = pd.DataFrame(dataset)
df_preprocessed = df_preprocessed.drop(['Unnamed: 0'],axis = 1)
pd.get_dummies(data=df_preprocessed,columns=['state','location','type'])
```

```python
#Checking for outliers
c = ['so2','no2','rspm','spm']
plt.figure(figsize=(10,5))
for i in range(0,len(c)):
    plt.subplot(5,1,i+1)
    sns.boxplot(x=dataset[c[i]],color='green',fliersize=5,orient='h')
    plt.tight_layout()
```

```python
# All the feature columns have outliers
# barplot
fig, axes= plt.subplots(figsize=(20, 12), ncols=4)
state_wise_max_so2 = dataset[['state','so2']].dropna().groupby('state').median().sort_values(by='so2')
state_wise_max_no2 = dataset[['state','no2']].dropna().groupby('state').median().sort_values(by='no2')
state_wise_max_rspm = dataset[['state','rspm']].dropna().groupby('state').median().sort_values(by='rspm')
state_wise_max_spm = dataset[['state','spm']].dropna().groupby('state').median().sort_values(by='spm')


sns.barplot(x='so2', y=state_wise_max_so2.index, data=state_wise_max_so2, ax=axes[0])
axes[0].set_title("Average so2 observed in a state")

sns.barplot(x='no2', y=state_wise_max_no2.index, data=state_wise_max_no2, ax=axes[1])
axes[1].set_title("Average no2 observed in a state")

sns.barplot(x='rspm', y=state_wise_max_rspm.index, data=state_wise_max_rspm, ax=axes[2])
axes[2].set_title("Average rspm observed in a state")

sns.barplot(x='spm', y=state_wise_max_spm.index, data=state_wise_max_spm, ax=axes[3])
axes[3].set_title("Average spm observed in a state")


plt.tight_layout()
```

```python
#Correlation using heatmap
sns.heatmap(dataset.corr(),annot=True,cmap= 'Blues')
```

```python
# Top 10 most pollutes states
st_wise = dataset.pivot_table(values=['so2','no2','rspm','spm'],index='state').fillna(0)
maxso2 = st_wise.sort_values(by='so2',ascending=False)
maxso2.loc[:,['so2']].head(10).plot(kind='bar')
```

```python
# Considering year from given date 
dataset['date'] = pd.to_datetime(dataset['date'])
dataset['year'] = dataset['date'].dt.year
dataset['year'].unique()
year_groupby = dataset.groupby(['year']).median().reset_index().sort_values(by = 'year', ascending = False)

#Variation of pollutants over year
# SO2
f,ax = plt.subplots(figsize = (10,7))
plt.xticks(rotation = 90)
sns.pointplot(x = 'year', y = 'so2', data = year_groupby, color= 'r')
plt.xlabel('Year')
plt.ylabel('So2')
plt.title('So2 emission trends over the years')
plt.show()
```

```python
# No2
f,ax = plt.subplots(figsize = (10,7))
plt.xticks(rotation = 90)
sns.pointplot(x = 'year', y = 'no2', data = year_groupby)
plt.xlabel('Year')
plt.ylabel('No2')
plt.title('No2 emission trends over the years')
plt.show()
```

```python
#rspm 
f,ax = plt.subplots(figsize = (10,7))
plt.xticks(rotation = 90)
sns.pointplot(x = 'year', y = 'rspm', data = year_groupby)
plt.xlabel('Year')
plt.ylabel('rspm')
plt.title('rspm emission trends over the years')
plt.show()
```

```python
#spm 
f,ax = plt.subplots(figsize = (10,7))
plt.xticks(rotation = 90)
sns.pointplot(x = 'year', y = 'spm', data = year_groupby)
plt.xlabel('Year')
plt.ylabel('spm')
plt.title('spm emission trends over the years')
plt.show()
```
